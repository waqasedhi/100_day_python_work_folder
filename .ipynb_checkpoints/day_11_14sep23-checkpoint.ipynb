```python
## main code
import random as rd

starter=input("Do you want to play a game of Blackjack? Type 'y' or 'n': ")
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

yourCards=[]
yourScore=0
comCards=[]
comScore=0
if starter == 'y':
    for i in range(2):
        card=0
        random1=rd.randint(0,12)
        card=cards[random1]
        if (card == 11) and (yourScore > 10):
            yourCards.append(1)
            yourScore+=1
        else:
            yourCards.append(card)
            yourScore+=card
            
    
    
    random1=rd.randint(0,12)
    comCards.append(cards[random1])
    comScore+=cards[random1]
    print(f"Your cards: {yourCards}, current score: {yourScore}")
    print(f"Computer's first cards: {comCards}")
    
#anotherCard=input("Type 'y' to get another card, type 'n' to pass: ")
    
continue1=True
while continue1:
    anotherCard=input("Type 'y' to get another card, type 'n' to pass: ")
    if anotherCard=='y':
        card=0
        random1=rd.randint(0,12)
        card=cards[random1]
        if (card == 11) and (yourScore > 10):
            yourCards.append(1)
            yourScore+=1
            if yourScore>21:
                while not comScore >= 16:
                    card2=0
                    random2=rd.randint(0,12)
                    card2=cards[random2]
                    if (card2 == 11) and (comScore > 10):
                        comCards.append(1)
                        comScore+=1
                    else:
                        comCards.append(card2)
                        comScore+=card2
                continue1=False
                print(f"Your cards: {yourCards}, current score: {yourScore}")
                print(f"Computer's first cards: {comCards}")
                print(f"Your final hand: {yourCards}, final score: {yourScore}")
                print(f"Computer final hand: {yourCards}, final score: {yourScore}")
                print("You went over. you lose")
            else:
                print(f"Your cards: {yourCards}, current score: {yourScore}")
                print(f"Computer's first cards: {comCards}")
        else:
            yourCards.append(card)
            yourScore+=card
            if yourScore>21:
                while not comScore >= 16:
                    card2=0
                    random2=rd.randint(0,12)
                    card2=cards[random2]
                    if (card2 == 11) and (comScore > 10):
                        comCards.append(1)
                        comScore+=1
                    else:
                        comCards.append(card2)
                        comScore+=card2
                continue1=False
                print(f"Your cards: {yourCards}, current score: {yourScore}")
                print(f"Computer's first cards: {comCards}")
                print(f"Your final hand: {yourCards}, final score: {yourScore}")
                print(f"Computer final hand: {yourCards}, final score: {yourScore}")
                print("You went over. you lose")
            else:
                print(f"Your cards: {yourCards}, current score: {yourScore}")
                print(f"Computer's first cards: {comCards}")

    else:
        
        continue1=False
```
